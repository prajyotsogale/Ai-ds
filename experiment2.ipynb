```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df= pd.read_csv("futuristic_city_traffic (1).csv")
df2 = df
```

```python
columns = list(df.columns)
print(columns)
```

```python
print(plt.figure(figsize=(10, 4), dpi=100))
sns.boxplot(x="Hour Of Day", data=df2)
```

```python
plt.figure(figsize=(10,4),dpi = 100)
color_palette = sns.color_palette("Accent_r")
sns.set_palette(color_palette)
sns.countplot(x="Hour Of Day", data=df2)
```

```python
plt.figure(figsize=(10,4),dpi=100)
color_palette = sns.color_palette("cool")
sns.set_palette(color_palette)
sns.countplot(x="Day Of Week", data=df2)
```

```python
grp = dict(df2.groupby('Economic Condition').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
        plt.title("economic condition") 
        plt.pie(m.values(), labels = m.keys())
```

```python
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "Economic Condition", data = df2) 
```

```python
# Region analysis of different bmi
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("magma") 
sns.set_palette(color_palette)
sns.barplot(x = "Economic Condition", y = "Is Peak Hour", data = df2) 
```

```python
# Region analysis as per charges 
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
 
sns.barplot(x = "Economic Condition", y = "Is Peak Hour", data = df2) 
```

```python
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
sns.barplot(x = "Economic Condition", y = "Is Peak Hour", data = df2) 
```

```python
# Finding correlation 
plt.figure(figsize = (10, 6), dpi = 100) 
color = sns.color_palette("magma")
sns.heatmap(df2.corr(), vmax=0.9, annot=True,cmap = color)
```
